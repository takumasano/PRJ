```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib.colors import Normalize
from scipy.io import loadmat

data = loadmat('/Users/tstakuma/Desktop/PRJ/samples/sample data/1_19-Oct-2019_10-18-23_mouse.mat')

# Extract x and y data from the loaded data
x = data['mouseMovement']  # Replace 'x' with the actual variable name in your .mat file
y = data['mouseMovement']  # Replace 'y' with the actual variable name in your .mat file

def traj_withColour(x, y, fig=None, ax=None):
    if fig is None:
        fig, ax = plt.subplots()
        colors = np.linspace(0, 1, len(x))
        ax.plot(x, y, '-k', alpha=0.2)
        ax.scatter(x, y, c=colors, cmap=cm.turbo)
        ax.plot(x[0], y[0], 'Dr', label='start', markersize=8)
        ax.axis('equal')
        norm = Normalize(vmin=0, vmax=len(x))
        cbar = fig.colorbar(cm.ScalarMappable(cmap=cm.turbo, norm=norm), ax=ax)
        cbar.set_label('Time step')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.legend()
        return fig, ax  
    
traj_withColour(x, y)
# Show the plot
plt.show()
```
